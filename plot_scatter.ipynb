# Plot scatter graph

* data set : pandas form (csv->pandas)
* font : Times new Roman

## 1. Load dataset

````python
import pandas as pd
data = pd.read_csv('./python/cbt5_log/0_100_day.csv')
data.head()
````

Dataset has 23 columns and every columns except churn and label are numeric.


## 2. Plot the chart

1) Split dataframe by specific column.

2) Plot it by using scatter.

````python
import matplotlib.pyplot as plt
import matplotlib
````

````python
# First, split data by 'label' column.
data_all = data.loc[data['label']=='allOK']
data_soc = data.loc[data['label']=='socOK']
data_ind = data.loc[data['label']=='indOK']
data_no = data.loc[data['label']=='noOK']
````

We can select data by indexing, Example is ```ax.scatter(data_all[X_column],data_all[Y_column],color='r',alpha=.4,label='first_label')```

First 2 elements are for X axis and Y axis, **color** is for coloring (i.e. r,g,b,gray,yellow and etc), **alpha** is for transparency and **label** is for legend.

````python
fig, ax = plt.subplots()
ax.scatter(data_all['die_num'],data_all['highlevel'],color='r',alpha=.4,label='acculturation')
ax.scatter(data_soc['die_num'],data_soc['highlevel'],color='g',alpha=.4,label='assimilation')
ax.scatter(data_ind['die_num'],data_ind['highlevel'],color='b',alpha=.4,label='separation')
ax.scatter(data_no['die_num'],data_no['highlevel'],color='black',alpha=.4,label='marginality')
## Make a legend at upper right
plt.legend(loc='upper right')
## X, Y label name
plt.xlabel('die #')
plt.ylabel('level')
plt.show()
````

### 2) Change Font

For same code, Let's change the font of X, Y labels and title.

I will use Times New Roman font. Below code makes Every matplotlib font-type to 'Times New Roman' and size to 15.

````python
import matplotlib as mpl
mpl.rc('font',family='serif',serif='Times New Roman', size=15)
````

````python
fig, ax = plt.subplots()
ax.scatter(data_all['die_num'],data_all['highlevel'],color='r',alpha=.4,label='acculturation')
ax.scatter(data_soc['die_num'],data_soc['highlevel'],color='g',alpha=.4,label='assimilation')
ax.scatter(data_ind['die_num'],data_ind['highlevel'],color='b',alpha=.4,label='separation')
ax.scatter(data_no['die_num'],data_no['highlevel'],color='black',alpha=.4,label='marginality')
## Make a legend at upper right
plt.legend(loc='upper right')
## X, Y label name
plt.title('Example of Graph')
plt.xlabel('die #')
plt.ylabel('level')
plt.show()
````

### 3) Markers

There are lots of styles for markers.
The details in https://matplotlib.org/api/markers_api.html#module-matplotlib.markers.

Let's change all markers to circle, square, triangle_up, triangle_down

````python
fig, ax = plt.subplots()
ax.scatter(data_all['die_num'],data_all['highlevel'],color='r',alpha=.4,marker="o",label='acculturation')
ax.scatter(data_soc['die_num'],data_soc['highlevel'],color='g',alpha=.4,marker="s",label='assimilation')
ax.scatter(data_ind['die_num'],data_ind['highlevel'],color='b',alpha=.4,marker="^",label='separation')
ax.scatter(data_no['die_num'],data_no['highlevel'],color='black',alpha=.4,marker="v",label='marginality')
## Make a legend at upper right
plt.legend(loc='upper right')
## X, Y label name
plt.title('Example of Graph')
plt.xlabel('die #')
plt.ylabel('level')
plt.show()
````
