## power spectrum

```python

from numpy import sin, linspace, pi
import matplotlib.pyplot as plt
from scipy import fft, arange

def plotSpectrum(y,Fs):

    n = len(y) # length of the signal
    k = arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range

    frq = frq[range(int(n/2))] # one side frequency range

    Y = fft(y)/n # fft computing and normalization
    Y = Y[range(int(n/2))]

#     plt.plot(frq,abs(Y),'r') # plotting the spectrum
#     plt.xlabel('Freq (Hz)')
#     plt.ylabel('|Y(freq)|')
    return frq, abs(Y)

# per 1 hours. 30days. 
def undersampling(data,min_unit ):
#     max_dates = data[data.columns[0]].max()
#     min_dates = data[data.columns[0]].min()
    hour = int(60/min_unit)
    all_data = []
    count = 0
    for idx,da in enumerate(data):
        if idx%hour==0:
            count+=1
            all_data.append(da)
        if count>=168:
            break
    return all_data
```

```python
taxi = pd.read_csv('./dataset/NAB/realKnownCause/nyc_taxi.csv')
taxi.head()
```

```python

# data_temp = undersampling(data[data.columns[1]],5)
data_temp = undersampling(taxi.value,30)

Fs = len(data_temp)  # sampling rate
Ts = 1.0/Fs # sampling interval
t = arange(0,1,Ts) # time vector

fig = plt.figure(figsize=(8,6))
y=data_temp

ax1 = fig.add_subplot(2,1,1)
ax1.grid(alpha=.4,linewidth=.5,linestyle='--')
ax1.plot(y,color='b',linewidth='1')
ax1.yaxis.set_major_formatter(tick) 
plt.xticks()
plt.xlabel('time (1 hour)')
plt.ylabel('value')
plt.title('Real value')

X_1,Y_1 = plotSpectrum(y,Fs)
ax2 = fig.add_subplot(2,1,2)
ax2.grid(alpha=.4,linewidth=.5,linestyle='--')
ax2.plot(X_1[:25],Y_1[:25],color='r',linewidth='1')
ax2.scatter(X_1[7],Y_1[7],c='#000000')
ax2.yaxis.set_major_formatter(tick) 
plt.xlabel('Frequency')
plt.xticks(range(0,25))
plt.ylabel('power')
plt.title('Power spectral density')

fig.tight_layout()
plt.show()
fig.savefig('./paper/fig6.eps', format='eps', dpi=1000)
```
